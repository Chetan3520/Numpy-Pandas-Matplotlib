### Adjusting the plot:Lines and Colors Style

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)(Darkyellow)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1 (violate/Red)
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported (faint green)



```

#### matplotlib library have functionality that take bydefault colors of the lines in plots.

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

# Barchart

### what is Barchart?

##### A bar plot is plot which represent the categorical data and there corresponding frequencies with rectangular bars length proportional to count(frequencies) that they presents.

### When use?

##### 1)To represent the frequency distribution of the categorical variables.

##### 2)A bar diagrame makes it easy to sets of data between different groups.

#####  3) To compare multiple categorical variables at a one time with their corresponding frequencies.



```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
company=["Google","Amazon","Microsoft","Facebook"]
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue,label="Revenue")
plt.legend()
plt.ylabel("revenue($bln)")
plt.title("US Tech Stocks")
```

### Also want to show each companies Profit side by side.

```python
plt.bar(company,revenue,label="Revenue")
plt.ylabel("revenue($bln)")
plt.title("US Tech Stocks")
plt.bar(company,profit,label="profit")
plt.legend()
```

## we want to generate side by side plot of both revenue and profit so,

### we just have to change the x axis string values into int, for that,we will convert to companies into array of integer values that will lead to be easy for further operations.

```python
xpos=np.arange(len(company))
```

```python
plt.bar(xpos,revenue,width=0.3,label="revenue")
plt.bar(xpos,profit,width=0.3,label="profit")
plt.legend()
```

### we have to change the positions of the plot for profit so we add some value in the xpos for profit bars

```python
plt.bar(xpos,revenue,width=0.3,label="revenue")
plt.bar(xpos+0.3,profit,width=0.3,label="profit")
plt.legend()
```

### but we have to again give the names of companies that on x axis for that

```python
plt.bar(xpos,revenue,width=0.3,label="revenue")
plt.bar(xpos+0.3,profit,width=0.3,label="profit")
plt.xticks(xpos,company)
plt.legend()
plt.title("US tech companies Profit and Revenues")
plt.xlabel("US Tech")
plt.ylabel("Revenue and Profit")
plt.grid()
```

## To Roatade Barplot Horizonatal plt.barh()

```python
plt.yticks(xpos,company)
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,profit,label="profit")
plt.title("US tech companies Profit and Revenues")
plt.xlabel("US Tech")
plt.ylabel("Revenue and Profit")
plt.legend()
plt.grid()
```

# Histogram

### What is the Histogram?

##### It is graphical representation of the data using bars of the different hieght.

##### Frequency distribution of the numerical data

##### Histograme ,groups the numbers into ranges and the hieght of the each bar depicts the frequency of the each range or bin.

### When to use?

##### Represent the frequency distribution of the numerical data.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
#80-100:Normal
#100-120:pre-diabetic
#125 Above:Diabetic
```

#### we have to find the Ans:

```python
#1)How many patience are normal?
#2)How many of them are pre-diabetic?
#3)How many are diabetic?


```

```python
plt.hist(blood_sugar)
```

#### if we want to increase or decrease the bins and corresponding bars.

```python
plt.hist(blood_sugar,bins=3)
```

#### We want bins(range) of the the bars that ans the Questions.

```python
plt.hist(blood_sugar,bins=[80,100,120,125,150],rwidth=0.95,color="green")
#rwidth: Relative width of the bars compare to a bin size.
```

#### so we can say that,

```python
#1)How many patience are normal?   
#ans: 6
#2)How many of them are pre-diabetic?
#ans: 2
#3)How many are diabetic?
#ans: 4

```

## Side by Side two Histogram

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
len(blood_sugar_men)==len(blood_sugar_women)
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],color=["green","orange"],label=["men","women"])
plt.xlabel("Sugar Range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.legend()
plt.grid()
```

### Plot Histogram Horizontal way

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],color=["green","orange"],label=["men","women"],
        orientation="horizontal")
plt.xlabel("Sugar Range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.legend()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
section=["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
exp=[1400,600,300,420,250]


```

```python
plt.axis("equal")

plt.pie(exp,labels=section)
```

### to give the percentages and increase the size of  the pie chart

```python
plt.pie(exp,labels=section,radius=1.5,autopct="%0.2f%%")#two digits the after decimal points.
plt.show()#to avoid the extra labels in the output
```

### to seperate the sections of the pie chart

```python
plt.pie(exp,labels=section,radius=1.5,autopct="%0.2f%%",explode=[0,0,0,0.2,0]) #seperate the fourth part of the pie.
plt.show()
#autopct= "%0.2%%":for percentages in the pie chart 2 digits after decimal points.
```

```python
plt.pie(exp,labels=section,radius=1.5,autopct="%0.2f%%",explode=[0,0,0,0.2,0],startangle=90) 
#use the startangle for roatating the pie chart.
plt.show()
```
